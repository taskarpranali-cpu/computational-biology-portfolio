## 2. Lists — Ordered Collections

Lists hold multiple items in order — like a rack of samples.

```python
# A list of gene names
cancer_genes = ["TP53", "EGFR", "BRCA1", "BRCA2", "KRAS", "MYC", "PIK3CA"]

# Accessing items (indexing starts at 0, like floors in a building)
print(f"First gene:  {cancer_genes[0]}")
print(f"Third gene:  {cancer_genes[2]}")
print(f"Last gene:   {cancer_genes[-1]}")

# How many genes?
print(f"Total genes: {len(cancer_genes)}")

# Add a gene
cancer_genes.append("PTEN")
print(f"After adding PTEN: {cancer_genes}")
```

```python
# A list of expression values (one per gene)
expression_values = [2.1, 8.5, 4.3, 3.9, 12.1, 9.7, 5.5, 1.8]

# Basic statistics without any libraries!
total = sum(expression_values)
count = len(expression_values)
average = total / count
maximum = max(expression_values)
minimum = min(expression_values)

print(f"Average expression: {average:.2f}")
print(f"Range: {minimum} to {maximum}")
```

## 3. Dictionaries — Key-Value Pairs

Dictionaries map keys to values — like a gene annotation table.

```python
# Map gene names to their functions
gene_functions = {
    "TP53":   "Tumor suppressor, cell cycle arrest",
    "EGFR":   "Growth factor receptor, cell proliferation",
    "BRCA1":  "DNA repair, homologous recombination",
    "KRAS":   "Signal transduction, GTPase",
}

# Look up a gene
print(f"TP53 function: {gene_functions['TP53']}")

# Map gene names to expression values
expression = {
    "TP53": 2.1,
    "EGFR": 8.5,
    "BRCA1": 4.3,
    "KRAS": 12.1,
}

# Which genes are highly expressed?
for gene, value in expression.items():
    if value > 5.0:
        print(f"  {gene}: {value} (HIGH)")
```

## 4. Loops & Conditionals — Automating Analysis

Loops let you repeat actions. Conditionals let you make decisions. Together, they automate analysis you'd otherwise do by hand.

```python
# Classify expression levels
genes = ["TP53", "EGFR", "BRCA1", "KRAS", "MYC"]
values = [2.1, 8.5, 4.3, 12.1, 9.7]

print("Gene Expression Classification")
print("-" * 35)

for i in range(len(genes)):
    gene = genes[i]
    val = values[i]
    
    # Classify based on thresholds
    if val > 10:
        category = "VERY HIGH"
    elif val > 5:
        category = "HIGH"
    elif val > 2:
        category = "MODERATE"
    else:
        category = "LOW"
    
    print(f"  {gene:8s}  {val:6.1f}  →  {category}")
```

```python
# A cleaner way: zip() pairs up two lists
for gene, val in zip(genes, values):
    status = "upregulated" if val > 5 else "normal"
    print(f"{gene}: {status}")
```

## 5. Functions — Reusable Biology Tools

Functions are like protocols in the lab — defined once, used many times.

```python
def gc_content(sequence):
    """Calculate the GC content of a DNA sequence.
    
    GC content = (G + C) / total length * 100
    
    Parameters:
        sequence (str): A DNA sequence string (A, T, G, C)
    
    Returns:
        float: GC content as a percentage
    """
    sequence = sequence.upper()  # Handle lowercase input
    g_count = sequence.count('G')
    c_count = sequence.count('C')
    total = len(sequence)
    
    if total == 0:
        return 0.0
    
    return (g_count + c_count) / total * 100


# Test it!
test_seq = "ATGCGATCGATCGATCG"
print(f"Sequence: {test_seq}")
print(f"GC content: {gc_content(test_seq):.1f}%")
```

```python
def reverse_complement(sequence):
    """Return the reverse complement of a DNA sequence."""
    complement = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}
    return ''.join(complement[base] for base in reversed(sequence.upper()))


def transcribe(dna_sequence):
    """Transcribe DNA to mRNA (replace T with U)."""
    return dna_sequence.upper().replace('T', 'U')


# Test both functions
dna = "ATGCGATCG"
print(f"DNA:                {dna}")
print(f"Reverse complement: {reverse_complement(dna)}")
print(f"mRNA:               {transcribe(dna)}")
```

## 6. MILESTONE: GC Content Analyzer

Time to put it all together! Compute GC content for multiple DNA sequences.

```python
# 10 DNA sequences (simulating reads from different genes)
sequences = {
    "BRCA1_exon1":  "ATGGATTTATCTGCTCTTCGCGTTGAAGAAGTACAAAATGTC",
    "TP53_exon4":   "TCCCCCTTGCCGTCTGGGCTTCTTGCATTCTGGGACAGCCAAGTCTGTG",
    "EGFR_exon18":  "AACATCTCCGAAAGCCAACAAGGAAATCCTCGATGTGAGTTTCTGCTTTGC",
    "KRAS_exon2":   "ATGACTGAATATAAACTTGTGGTAGTTGGAGCTGGTGGCGTAGGCAAGAGT",
    "MYC_exon2":    "ATGCCCCTCAACGTTAGCTTCACCAACAGGAACTATGACCTCGACTACGAC",
    "PTEN_exon5":   "AAAGCTGGAAAGGGACGAACTGGTGTAATGATATGTGCATATTTATTAC",
    "PIK3CA_ex9":   "AATGATGCTTGGCTCTGGAATGCCAGAACTACAATCTTTTGATGACAG",
    "BRAF_exon15":  "TGAAGACCTCACAGTAAAAATAGGTGATTTTGGTCTAGCTACAGTGAAATC",
    "APC_exon15":   "GAAATAGGATGTAATCAGACGACACAGGAAGCAGATTCTGCTAATACCCTG",
    "RB1_exon1":    "ATGCCGCCCAAAACCCCCCGAAAAACGGCCGCCACCGCCGCCGCTGCCGCC",
}

# Analyze each sequence
print(f"{'Gene':<16} {'Length':>6} {'GC%':>7}  Classification")
print("=" * 52)

gc_values = []

for name, seq in sequences.items():
    gc = gc_content(seq)
    gc_values.append(gc)
    
    # Classify GC content
    if gc > 60:
        label = "GC-rich"
    elif gc < 40:
        label = "AT-rich"
    else:
        label = "Balanced"
    
    print(f"{name:<16} {len(seq):>6} {gc:>6.1f}%  {label}")

print("=" * 52)
print(f"{'Average':<16} {'':>6} {sum(gc_values)/len(gc_values):>6.1f}%")
print(f"\nGC content range: {min(gc_values):.1f}% to {max(gc_values):.1f}%")
```
